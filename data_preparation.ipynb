```python
final_data_set_columns = list(infant_mortality_features(lines[0]).keys())

final_data_set_columns.append('T')
final_data_set_columns.append('Y')

data = []
for line in lines:
    data_row = infant_mortality_features(line)
    if line[569-1] == 'Y':
        data_row['T'] = 1
    elif line[569-1] == 'N':
        data_row['T'] = 0
    else: # unknown
        data_row['T'] = -1
    data_row['Y'] = 1 if line[1356-1:1358] != ' '*3 else 0
    data.append(data_row)

final_data_df = pd.DataFrame(data, columns=final_data_set_columns)
final_data_df.to_csv('data/infant_mortality_data.csv', index=False)
```

```python
def decode_RACE31(race_val, prefix):

    new_cols_lst = [f'{prefix}_White', f'{prefix}_Black', f'{prefix}_AIAN', f'{prefix}_Asian', f'{prefix}_NHOPI']
    vals = dict(zip(new_cols_lst, [0]*5))
    if race_val == '01':
        vals[f'{prefix}_White'] = 1
    elif race_val == '02':
        vals[f'{prefix}_Black'] = 1
    elif race_val == '03':
        vals[f'{prefix}_AIAN'] = 1
    elif race_val == '04':
        vals[f'{prefix}_Asian'] = 1
    elif race_val == '05':
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '06':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '07':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
    elif race_val == '08':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_Asian'] = 1
    elif race_val == '09':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '10':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '11':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
    elif race_val == '12':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '13':
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '14':
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '15':
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '16':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '17':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
    elif race_val == '18':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '19':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '20':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '21':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '22':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '23':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '24':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '25':
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '26':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '27':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
    elif race_val == '28':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '29':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '30':
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1
    elif race_val == '31':
        vals[f'{prefix}_Black'] = 1
        vals[f'{prefix}_AIAN'] = 1
        vals[f'{prefix}_Asian'] = 1
        vals[f'{prefix}_NHOPI'] = 1
        vals[f'{prefix}_White'] = 1

    return vals.values()

new_cols_lst_mother = ['M_White', 'M_Black', 'M_AIAN', 'M_Asian', 'M_NHOPI']
new_cols_lst_father = ['F_White', 'F_Black', 'F_AIAN', 'F_Asian', 'F_NHOPI']

new_values_mother = sampled_infant_mortality_data['MRACE31'].apply(lambda x: decode_RACE31(x, 'M'))
new_values_father = sampled_infant_mortality_data['FRACE31'].apply(lambda x: decode_RACE31(x, 'F'))

new_values_df_mother = pd.DataFrame(list(new_values_mother), columns=new_cols_lst_mother)
new_values_df_father = pd.DataFrame(list(new_values_father), columns=new_cols_lst_father)

if new_cols_lst_mother[0] not in sampled_infant_mortality_data.columns:
    # Concatenate the new columns with the original DataFrame
    sampled_infant_mortality_data = pd.concat([sampled_infant_mortality_data.reset_index(drop=True), new_values_df_mother], axis=1)

if new_cols_lst_father[0] not in sampled_infant_mortality_data.columns:
    # Concatenate the new columns with the original DataFrame
    sampled_infant_mortality_data = pd.concat([sampled_infant_mortality_data.reset_index(drop=True), new_values_df_father], axis=1)

display(sampled_infant_mortality_data.head(3))
```

```python
def numeric_null_detection(val):
    if val == '99' or val == '999' or val == '9999' or val == '99.9':
        return None
    return val

for col_name, col_type in types_dict.items():
    print(f"doing: {col_name}")
    print(f"columns len: {len(sampled_infant_mortality_data.columns)}")
    if col_name in sampled_infant_mortality_data.columns:
        if col_type == 'bool':
            sampled_infant_mortality_data[col_name] = sampled_infant_mortality_data[col_name].map({'Y': 1, 'N': 0, 'U': None})
        elif col_type == 'numeric':
            sampled_infant_mortality_data[col_name] = sampled_infant_mortality_data[col_name].apply(numeric_null_detection)
            sampled_infant_mortality_data[col_name] = pd.to_numeric(sampled_infant_mortality_data[col_name], errors='coerce')
        elif col_type == 'category':
            df_encoded = pd.get_dummies(sampled_infant_mortality_data[col_name], prefix=col_name)
            sampled_infant_mortality_data = pd.concat([sampled_infant_mortality_data, df_encoded], axis=1)
            sampled_infant_mortality_data.drop(columns=[col_name], inplace=True)
        print(f"columns len: {len(sampled_infant_mortality_data.columns)}")
```
