```python
# plot histogram of average smoking during pregnancy
import numpy as np

def to_num(cig_val):
    if cig_val == '00' or cig_val == '  ' or cig_val == '99':
        return 0
    if cig_val[0] == '0':
        return int(cig_val[1])

    return int(cig_val)
    

vals_cig_0 = [to_num(line[253-1:254]) for line in dead_lines]
vals_cig_1 = [to_num(line[255-1:256]) for line in dead_lines]
vals_cig_2 = [to_num(line[257-1:258]) for line in dead_lines]
vals_cig_3 = [to_num(line[259-1:260]) for line in dead_lines]

print("vals_cig_0:", vals_cig_0[:10])
vals_cig_avrg = (np.array(vals_cig_0) + np.array(vals_cig_1) + np.array(vals_cig_2) + np.array(vals_cig_3)) / 4
print(len(vals_cig_avrg))
plt.figure(figsize=(10, 6))
sns.histplot(vals_cig_avrg, bins=30, kde=True, stat='probability')
# normalize y vals

plt.xlabel('Average Smoking During Pregnancy')
plt.ylabel('Density')
plt.ylim(0, 1)  # Adjust y-axis limits for better visibility
plt.title('Histogram of Average Smoking During Pregnancy for Dead Individuals')
plt.tight_layout()
plt.savefig('average_smoking_dead_histogram.png')
plt.show()
```

```python
import pandas as pd

def for_latex(text):
    """Escape LaTeX special characters in a string."""
    return text.replace('_', '\\_').replace('%', '\\%').replace('&', '\\&').replace('$', '\\$').replace('<', '\\textless ').replace('>', '\\textgreater ')

def creat_latex_table_from_features_dictionary(features_dictionary, colors):

    new_features_dictionary = {}
    for category, features in features_dictionary.items():
        new_features_dictionary[category] = [(feature, description) for feature, description in features.items()]

    latex_table = "\\begin{table}[h!]\n"
    latex_table += "\\centering\n"
    latex_table += "\\caption{Features Dictionary for Infant Mortality Analysis}\n"
    latex_table += "\\label{tab:features_dictionary}\n"
    latex_table += "\\small\n"
    
    latex_table += f"\\begin{{tabular}}{{{'l'* len(new_features_dictionary.keys())}}}\n"

    latex_table += "\\hline\n"
    for key in new_features_dictionary.keys():
        latex_table += f"\\textbf{{{for_latex(key)}}} & "
    
    latex_table = latex_table[:-2] + "\\\\\n"  # Remove last '&' and add new line
    latex_table += "\\hline\n"

    longest_column = max(len(new_features_dictionary[key]) for key in new_features_dictionary.keys())
    for i in range(longest_column):
        for key in new_features_dictionary.keys():
            if i < len(new_features_dictionary[key]):
                latex_table += f"{for_latex(new_features_dictionary[key][i][0])} -  {for_latex(new_features_dictionary[key][i][1])}& "
            else:
                latex_table += " & "
        latex_table = latex_table[:-2] + "\\\\\n"  # Remove last '&' and add new line
    
    latex_table += "\\hline\n"
    latex_table += "\\end{tabular}\n"
    latex_table += "\\end{table}\n"
    return latex_table

# seperate the features dictionary into n features dictionary for easier print

max_features_per_table = 2
def split_features_dictionary(features_dict, max_features):
    """Split the features dictionary into smaller dictionaries with a maximum number of features."""
    split_dicts = []
    current_dict = {}
    for category, features in features_dict.items():
        if len(current_dict) >= max_features:
            split_dicts.append(current_dict)
            current_dict = {}
        current_dict[category] = features
    if current_dict:
        split_dicts.append(current_dict)
    return split_dicts

def create_latex_tables(features_dict, max_features, save_path= "tabel_features.tex"):
    """Create LaTeX tables from the features dictionary."""
    split_dicts = split_features_dictionary(features_dict, max_features)
    tex_tabels = ""
    for i, sub_dict in enumerate(split_dicts):
        tex_tabels += creat_latex_table_from_features_dictionary(sub_dict)
        tex_tabels +="\n"
    with open(save_path, 'w') as f:
        f.write(tex_tabels)

# print(creat_latex_table_from_features_dictionary(split_features_dictionary(features_dictionary, max_features_per_table)[0]))
create_latex_tables(features_dictionary, max_features_per_table, save_path="tabel_features.tex")
```
