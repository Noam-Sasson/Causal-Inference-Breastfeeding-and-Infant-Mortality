```python
import pandas as pd
```

```python
sampled_infant_mortality_data = pd.read_csv("sampled_infant_mortality_data.csv")
```

```python
display(sampled_infant_mortality_data.head(3))
```

```python
count = 0
for col in sampled_infant_mortality_data.columns:
    # print(f"Missing values in {col}: {sampled_infant_mortality_data[col].isnull().sum()/len(sampled_infant_mortality_data)}")
    p = 0.05
    if sampled_infant_mortality_data[col].isnull().sum()/len(sampled_infant_mortality_data) > p:
        print(col)
        count += 1

print(f"Number of columns with more than {p*100}% missing values: {count}")
```
