## Method 1: Naive Imputation
numeric: mean <br>
categorical: mode

```python
def replace_nan(cur_val, new_val):
    # print(cur_val)
    if pd.isna(cur_val):
        return new_val
    return cur_val

null_columns = sampled_infant_mortality_data_m1.columns[sampled_infant_mortality_data_m1.isnull().any()].tolist()
for col in null_columns:
    if col in types_dict:  # Check if column exists in the dictionary
        dtype = types_dict[col]  # Get the data type
        print(f"Processing {col} as {dtype}")
        
        if dtype == 'numeric':
            col_mean = sampled_infant_mortality_data_m1[col].mean(skipna=True)  # Explicitly skip NaN values
            if not pd.isna(col_mean):  # Only proceed if we have a valid mean
                sampled_infant_mortality_data_m1[col] = sampled_infant_mortality_data_m1[col].apply(lambda x: replace_nan(x, col_mean))
            else:
                print(f"Warning: Could not calculate mean for {col} - all values are NaN")
                
        elif dtype == 'category':
            non_null_data = sampled_infant_mortality_data_m1[col].dropna()
            if len(non_null_data) > 0:  # Check if there are any non-null values
                col_mode = non_null_data.mode()[0]  # Drop NaN values before calculating mode
                sampled_infant_mortality_data_m1[col] = sampled_infant_mortality_data_m1[col].apply(lambda x: replace_nan(x, col_mode))
            else:
                print(f"Warning: Could not calculate mode for {col} - all values are NaN")
                
        elif dtype == 'bool':
            # For boolean columns, use the most frequent value (mode)
            non_null_data = sampled_infant_mortality_data_m1[col].dropna()
            print(non_null_data.apply(lambda x: int(x)).value_counts())
            if len(non_null_data) > 0:  # Check if there are any non-null values
                col_mode = non_null_data.mode()[0]  # Drop NaN values before calculating mode
                sampled_infant_mortality_data_m1[col] = sampled_infant_mortality_data_m1[col].apply(lambda x: replace_nan(x, col_mode))
            else:
                print(f"Warning: Could not calculate mode for {col} - all values are NaN")
```

create estimators for missing value imputation

```python
from sklearn.linear_model import LogisticRegression, LinearRegression

t_predictor = LogisticRegression(max_iter=1000, random_state=42)
f_age_predictor = LinearRegression()
```

T predictions

```python
X_train_T = sampled_infant_mortality_data_m3.drop(columns=['FAGECOMB'])
X_train_T = X_train_T.dropna()
y_train_T = X_train_T['T']
X_train_T = X_train_T.drop(columns=['T'])
```

```python
X_train_F = sampled_infant_mortality_data_m3.drop(columns=['T'])
X_train_F = X_train_F.dropna()
y_train_F = X_train_F['FAGECOMB']
X_train_F = X_train_F.drop(columns=['FAGECOMB'])
```

```python
# train models
t_predictor.fit(X_train_T, y_train_T)
f_age_predictor.fit(X_train_F, y_train_F)
```

```python
# predict for missing values

def predict_for_missing_val(cur_val, model, row, feature_cols):
    if pd.isna(cur_val):
        # Turn row slice into a DataFrame with one row
        row_features = pd.DataFrame([row[feature_cols].values], columns=feature_cols)
        predicted_val = model.predict(row_features)[0]
        return predicted_val
    return cur_val
```

```python
sampled_infant_mortality_data_m3['T'] = sampled_infant_mortality_data_m3.apply(
    lambda row: predict_for_missing_val(row['T'], t_predictor, row, X_train_T.columns),
    axis=1
)
sampled_infant_mortality_data_m3['FAGECOMB'] = sampled_infant_mortality_data_m3.apply(
    lambda row: predict_for_missing_val(row['FAGECOMB'], f_age_predictor, row, X_train_F.columns),
    axis=1
)
```
